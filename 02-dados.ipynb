Vamos começar carregando alguns dados. Para isso, vamos usar dados toy já contidos no scikit-learn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

iris = sns.load_dataset('iris')
```

```python
#print(iris.get_values)
iris.head()
#iris.describe()
```

Agora, vamos plotar algumas coisas, começando com um histograma dos valores do 'sepal length (cm)''.

```python
plt.hist(iris["sepal_length"], 10, density=True, facecolor='r', alpha=0.75)

plt.xlabel("Sepal Length")
plt.ylabel('Counting')
plt.title('Histogram')
plt.grid(True)
plt.show()
```

Usando o seaborn para ver melhor as distribuições por classe.

```python
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.show() 
```

E vamos ver as correlações entre os atributos.

```python
sns.pairplot(iris, kind="scatter", hue="species") # try kind="reg"
plt.show()
```

Começando a brincar com os dados. Vamos começar por separar a classe em um vetor e os atributos em uma matriz.

```python
y = iris.pop('species').values
#print(y)
list(iris)
x = iris.as_matrix()
```

```python
from sklearn import preprocessing

x_scaled = preprocessing.scale(x)

plt.hist(x_scaled[:,0], 10, density=True, facecolor='r', alpha=0.75)

plt.xlabel("Sepal Length")
plt.ylabel('Counting')
plt.title('Histogram')
plt.grid(True)
plt.show()
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)

plt.hist(x_scaled[:,0], 10, density=True, facecolor='r', alpha=0.75)

plt.xlabel("Sepal Length")
plt.ylabel('Counting')
plt.title('Histogram')
plt.grid(True)
plt.show()
```
